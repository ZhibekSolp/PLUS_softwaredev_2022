# Print a map with Matplotlib and Geopandas in Python

```python
import geopandas as gpd
```

```python
#Directory where the shapefile is hosted. Please download it folder Dataset

#"Dataset for Europe". Downloaded from http://tapiquen-sig.jimdo.com. 
#Based on shapes from Enviromental Systems Research Institute (ESRI). Free Distribuition.

shapefile = 'C:\\Users\\solpi\\repos\\a3_zhibek\\Dataset\\Europe\\Europe.shp'
europa_gdf = gpd.read_file(shapefile)
europa_gdf
```

```python
europa_gdf = europa_gdf.drop('ORGN_NAME', axis = 1)
europa_gdf
```

```python
europa_gdf.plot()
```

```python
import pandas as pd
#Directory where the shapefile is hosted.
percapita_df = pd.read_csv('C:\\Users\\solpi\\repos\\a3_zhibek\\Dataset\\percapita.csv')
```

```python
europa_gdf['percapita'] = percapita_df['percapita']
europa_gdf
```

```python
#choosing colormaps in Matplotlib
europa_gdf.plot(column = 'percapita', cmap = 'gist_ncar')
```

```python
import matplotlib.pyplot as plt
```

```python
#Create a map layout
fig = plt.figure(figsize=(9, 7))
ax = fig.add_subplot(111)

#Map title
title = 'GDP Per Capita (USD) of European countries'

#Title printing
plt.title(title + '\n', fontsize=20)

#Map printing
europa_gdf.plot(column='percapita',  cmap = 'gist_ncar', ax = ax)

#Name of the horizontal and vertical axes
ax.set_xlabel('Longitude', fontsize = 15)
ax.set_ylabel('Latitude', fontsize = 15)

#Definition of the minimum and maximum values as limits of the member to be used
min_p = min(europa_gdf['percapita'])
max_p = max(europa_gdf['percapita'])

#Definition of bar graph
bar = plt.cm.ScalarMappable(cmap='gist_ncar', norm=plt.Normalize(vmin = min_p , vmax=max_p))

#A list is declared where the values of the member will be stored.
bar._A = []
#Bar definition (location and size)
cax = plt.axes([0.85, 0.15, 0.03, 0.7])  #(x-coordinate, y-coordinate, width, length)

# Printout of the bar graph showing the upper variables 
cbar = fig.colorbar(bar, cax=cax)
#Bar title
cbar.set_label('Per Capita GDP (USD)', fontsize = 15)

#Saving the map in a jpg file
plt.savefig('Europa.jpg', dpi = 300)
```
